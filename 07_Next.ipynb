A couple of axes are worth noticing, because the advanced notebooks will stretch each of them:

- **On-policy vs off-policy.** REINFORCE, A2C and PPO train on fresh rollouts; DQN learns from a
  replay buffer. *Offline RL* takes the off-policy extreme — no new interaction at all.
- **Discrete vs continuous actions.** DQN is natural for discrete, PPO/A2C scale to continuous.
  Robotics and control problems live mostly on the continuous side.
- **Model-free vs model-based.** Everything above is model-free: the agent never tries to predict
  the environment. *Model-Based RL* and *Dreamer* drop that assumption and let the agent plan
  inside a learned simulator.
- **Single agent, single task.** The whole course so far trains one agent on one task. *MARL* and
  *Meta-RL* break both of those assumptions.

## What we haven't touched yet

If you list what the foundations *didn't* cover, you essentially get the table of contents for the
rest of the course:

- What if rewards are rare, or the agent has to discover them on its own? → **Exploration**
- What if there are other agents in the world — cooperating or competing? → **MARL**
- What if we can't interact with the environment at all, only a logged dataset? → **Offline RL**
- What if we want to plan ahead explicitly, like in chess or Go? → **MCTS / AlphaZero**
- What if the "environment" is a human, and the "agent" is a language model? → **RLHF / DPO / GRPO**
- What if the agent has a body, a camera, and a natural-language instruction? → **VLAs**
- How do we actually *run* all of this at scale, reliably? → **Productionisation**
- What if we want the agent to imagine instead of act? → **Model-Based RL** and **Dreamer**
- What if we need to adapt quickly to brand new tasks? → **Meta-RL**

Each of those questions gets its own notebook. Below we group them into themes to give you a
mental map before you dive in.

## The thematic map forward

Every advanced notebook is self-contained — pick whichever theme grabs you. But the numbering
isn't random: it traces a **five-act arc** that builds from relaxing the foundations' assumptions
all the way to the course's most ambitious algorithms. If you want a default path, just go in
order.

### Act 1 · Breaking the foundations' assumptions (08 → 10)

Every classic algorithm makes assumptions that keep the maths tractable. This act breaks them one
at a time and asks: *now what?*

- **`08_EXPL.ipynb` — Exploration & Curiosity.** Strip away dense reward and the agent flails.
  *Random Network Distillation (RND)* teaches it to chase novelty, cracking sparse-reward tasks
  like `MiniGrid-DoorKey-16x16` that are hopeless for vanilla PPO.
- **`09_MARL.ipynb` — Multi-Agent RL.** Drop the single-agent assumption and the environment
  becomes *non-stationary* — the other agents are learning too. You'll implement a
  centralised-training / decentralised-execution algorithm.
- **`10_OFFL.ipynb` — Offline RL.** Drop interaction itself. You have a logged dataset and must
  squeeze a policy out of it without ever touching the environment — grappling with
  *distributional shift*, the defining challenge of the field.

### Act 2 · Planning with a model (11)

What if the agent doesn't just *learn* but *plans*? The classic ideas that made AlphaGo work.

- **`11_MCTS.ipynb` — Monte Carlo Tree Search & AlphaZero.** Build MCTS from scratch on board
  games, then wire it to a deep network and self-play — the architecture behind AlphaZero.

### Act 3 · RL in the modern AI stack (12 → 13)

The two most visible RL applications of the last few years aren't about games at all — they're
about aligning language models and controlling physical robots.

- **`12_RLHF.ipynb` — RLHF, DPO & GRPO.** From classic PPO-based RLHF, through the simpler *DPO*,
  to the group-relative *GRPO* training modern reasoning models like DeepSeek R1 and OpenAI o3.
- **`13_DTVLA.ipynb` — Decision Transformers & VLAs.** Reframe RL as sequence modelling, then
  scale the idea up to Vision-Language-Action models — the class of model behind recent robot
  foundation systems like π0.

### Act 4 · From notebook to system (14)

A brief interlude on the plumbing that separates a training run you can share from a real
production pipeline.

- **`14_PROD.ipynb` — Productionising RL.** TensorBoard for experiment tracking, Ray for
  parallelism, Optuna for hyperparameter search.

### Act 5 · The model-based climax (15 → 16)

The final stretch — the most ambitious algorithms in the course, deliberately back-to-back because
they share a single idea: *learn the world, then train inside it.*

- **`15_MBRL.ipynb` — Model-Based RL.** Dyna-style algorithms à la *MBPO*: learn the dynamics,
  use them as a simulator, and get dramatic sample-efficiency wins over model-free baselines.
- **`16_DR3AM.ipynb` — Dreamer.** The state of the art. Learn a *latent* world model and train
  the whole policy inside imagined rollouts — the most complex algorithm you'll build in this
  course, and its capstone.

### Appendix · Learning to learn (17)

- **`17_META.ipynb` — Meta-Reinforcement Learning.** "Learning to learn": train on a distribution
  of tasks so the agent adapts to a new, unseen task in just a handful of episodes.
  Intellectually beautiful, mathematically dense, and — honestly — still niche in practice. A
  great intellectual diversion once you're done with the main track.

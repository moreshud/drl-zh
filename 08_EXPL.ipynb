# Exploration and Curiosity

In Reinforcement Learning (RL), agents learn by maximizing a **reward signal**. This core idea,
often called the _reward hypothesis_, means the design and frequency of rewards heavily influence
how well an agent learns.

However, we often face two major hurdles:

- **Difficult Reward Design:** Creating informative reward functions (extrinsic rewards from the
  environment) can be complex, especially for intricate tasks.
- **Sparse Rewards:** Many environments offer rewards only rarely. Imagine a maze where the agent
  only gets a reward (+1) at the exit and zero everywhere else. Without frequent feedback, the agent
  might wander aimlessly or get stuck, failing to explore effectively and find the goal.

When extrinsic rewards are scarce, standard exploration methods (like adding noise to actions) might
not be enough. The agent needs an internal drive to explore novel situations, a form of _curiosity_.
The overall reward is now given by:

$
r_t = e_t + i_t
$

I.e., the sum of the _extrinsic_ reward and an _intrinsic_ component.

This notebook explores [Random Network Distillation (RND)](https://arxiv.org/abs/1810.12894), a
powerful technique that provides the agent with such an intrinsic reward. RND achieved SoTA in
Montezuma's Revenge, a game famously difficult for deep reinforcement learning methods.

### Training Loop


```python
def train(
    env: gym.vector.VectorEnv, agent: BaseAgent, rollout_size=16, log_every=100, solved_score=0.95
):
    """The A2C training loop. We use the utility BaseAgent class for on-policy policy-gradient
    algorithms just for convenience in this notebook."""
    n_step = 0
    avg_score = 0.0
    obs, _ = env.reset()
    obs = torch.Tensor(obs).to(DEVICE)
    while True:
        # Collects a trajectory segment across multiple vectorized environments, making sure that
        # we take intrinsic rewards into account. Feel free to check out the library implementation
        # of `collect_trajectory_segment` which is very similar to the one you implemented already
        # in the PPO notebook!
        segment = collect_trajectory_segment(
            env, agent, obs, rollout_size, with_intrinsic_values=True
        )
        # Make the agent learn via the just-collected trajectory.
        stats = agent.learn(segment)
        # Update the state... Easy to overlook!
        obs = segment.next_start_obs
        print(
            f'Loss: {stats["loss"]: .7f}, Instrinsic reward (avg): '
            + f'{stats["rnd_reward_avg"]: .7f}, RND Loss: {stats["rnd_loss"]: .7f}\r',
            end="",
        )

        n_step += 1
        if n_step % log_every == 0:
            avg_score = np.mean(env.return_queue)
            print(f"Global step: {n_step*rollout_size}, Average Score: {avg_score:.5f}".ljust(100))

        if avg_score > solved_score:
            print(f"Environment solved with avg_score: {avg_score:.5f}".ljust(100))
            break
```

## Random Network Distillation

RND encourages the agent to visit unfamiliar states. It measures novelty by seeing how accurately a
_predictor_ network can guess the output of a _fixed, randomly initialized target network_ given the
current state.

- **Familiar states:** The predictor network learns to accurately predict the target's output.
  Prediction error is low -> low intrinsic reward (boredom).
- **Novel states:** The predictor network struggles to predict the target's output. Prediction error
  is high -> high intrinsic reward (curiosity).

This intrinsic reward motivates the agent to explore parts of the environment it hasn't seen often,
helping it overcome sparse extrinsic rewards and learn more effectively. RND cleverly avoids issues
found in other intrinsic motivation methods (like getting distracted by unpredictable but irrelevant
things, e.g., a noisy TV screen).

Before plugging RND into A2C, a quick sanity check. If the module works, training the predictor on a
batch of "seen" observations should drive its error toward zero on that batch, while the error on a
disjoint batch of "novel" observations should stay high — that's the novelty signal the intrinsic
reward is supposed to provide.


```python
def test_rnd_module():
    """Sanity check: RND produces a higher intrinsic reward on *novel* observations than on
    observations whose target the predictor has been trained to match."""
    init_random(seed=42)
    rnd = RNDModule().to(DEVICE)
    opt = optim.Adam(rnd.predictor.parameters(), lr=1e-3)

    # Two disjoint batches of observations.
    seen = torch.rand(8, 3, 7, 7, device=DEVICE)
    novel = torch.rand(8, 3, 7, 7, device=DEVICE)

    # Baseline error on `seen`, measured with dropout disabled for consistency.
    rnd.eval()
    initial_seen_error = rnd(seen).mean().item()
    rnd.train()

    # Train the predictor to imitate the (frozen) target on `seen`.
    for _ in range(200):
        opt.zero_grad()
        loss = rnd(seen).mean()
        loss.backward()
        opt.step()

    rnd.eval()
    final_seen_error = rnd(seen).mean().item()
    final_novel_error = rnd(novel).mean().item()

    assert (
        final_seen_error < initial_seen_error
    ), f"Predictor didn't learn: seen error {initial_seen_error:.4f} -> {final_seen_error:.4f}"
    assert (
        final_novel_error > final_seen_error
    ), f"No novelty signal: novel {final_novel_error:.4f} <= seen {final_seen_error:.4f}"
    print(
        f"✅ RND novelty test passed "
        f"(seen: {initial_seen_error:.4f} -> {final_seen_error:.4f}, "
        f"novel: {final_novel_error:.4f})"
    )


test_rnd_module()
```

### A2C w/ RND

Let's now re-implement A2C integrating the intrinsic reward from RND. Two design choices are worth
calling out upfront:

**Two value heads.** Extrinsic and intrinsic rewards have very different statistics (sparse,
episodic `{0, 1}` vs. dense, non-episodic novelty), so we keep _separate_ value estimates $V_e(s)$
and $V_i(s)$ — each with its own discount factor and advantage computation. In the code, $A_e$
respects episode boundaries via the `dones` mask, while $A_i$ does _not_: curiosity is about
exploring novel states, and resetting the intrinsic return on episode termination would throw that
signal away.

**Combined advantage.** The policy gradient is driven by a weighted sum of the two advantages:

$$
A(s, a) = \eta_{\text{ext}}\, A_e(s, a) + \eta_{\text{int}}\, A_i(s, a)
$$

In this notebook we pick $\eta_{\text{ext}}=100$ and $\eta_{\text{int}}=10$, i.e., the extrinsic
signal is weighted **10x** the intrinsic one. The intuition: the intrinsic reward _bootstraps_
learning when extrinsic reward is sparse, but the task is still defined by the extrinsic reward — we
don't want curiosity to dominate forever. The absolute scale of the two $\eta$ mostly shows up as
the magnitude of the policy-gradient step, which is why we pair them with a tight
`max_grad_norm_rnd=0.01` to keep the RND branch stable. The original paper instead scales the
intrinsic reward by a _running standard deviation_ estimate, which removes the need to re-tune these
coefficients per environment; we skip that trick here for simplicity, but note it as the more robust
alternative.

## Frontiers

RND is one member of a broader family of **exploration bonuses**. A few directions worth exploring
next:

- **[Intrinsic Curiosity Module (ICM)](https://arxiv.org/abs/1705.05363).** Instead of predicting a
  random target, ICM learns a forward model of next-state features and uses its prediction error as
  the intrinsic reward. ICM is task-coupled (the feature encoder is trained via an inverse model,
  focusing it on aspects of the environment the agent can actually influence) — but it can still be
  fooled by irrelevant stochasticity (the "noisy-TV" problem), which is exactly what RND was
  designed to fix.
- **Count-based exploration.** Classic tabular MDPs use visit counts $N(s)$ and a bonus like
  $\beta / \sqrt{N(s)}$. Neural variants approximate counts via density models
  ([Bellemare et al. 2016](https://arxiv.org/abs/1606.01868)) or hashing
  ([Tang et al. 2017](https://arxiv.org/abs/1611.04717)).
- **[NoisyNets](https://arxiv.org/abs/1706.10295).** Inject learnable parameter-space noise into the
  policy network itself, providing a cheap form of state-dependent exploration without a separate
  curiosity module.
- **[Go-Explore](https://arxiv.org/abs/1901.10995).** Skip bonuses altogether: remember promising
  states and _teleport back_ to them before exploring further. This return-then-explore scheme
  dramatically outperformed RND on Montezuma's Revenge.
- **Production-grade RND.** Two details we glossed over that matter at scale: (1) **observation
  normalization** for the target/predictor networks — without it, the predictor can trivially match
  the target on nearly-constant inputs; and (2) **scaling intrinsic rewards by a running std** so
  the combined-advantage coefficients don't need per-environment re-tuning. See the
  [OpenAI reference implementation](https://github.com/openai/random-network-distillation).

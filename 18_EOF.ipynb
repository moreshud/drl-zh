## Where the frontier is going

The course is finite; the field is not. Here is a short, opinionated tour of directions that feel
especially live right now — things we didn't cover in enough depth (or at all), and that are worth
your attention over the next year or two.

### RL for reasoning in LLMs

The single biggest RL story of the last couple of years has been *RL for reasoning*. Models like
OpenAI's **o1** / **o3** and DeepSeek's **R1** are trained with RL objectives (often variants of
the GRPO you implemented in `12_RLHF.ipynb`) on problems with verifiable answers — maths, code,
proofs. The reward signal is essentially "did the final answer check out", and long chains of
thought emerge from optimising against it. This has turned RL from a niche of game-playing and
robotics into one of the core training recipes for frontier LLMs, and the open questions (process
rewards vs outcome rewards, exploration in token space, avoiding reward hacking) are wide open.

### World models at scale

`16_DR3AM.ipynb` gave you a taste of what happens when an agent trains inside a learned world
model. The next step is *generative, interactive* world models big enough to serve as general
training grounds. **Genie** and **Genie 2** (DeepMind) generate playable 2D and 3D environments
from a single image; **V-JEPA** learns predictive video representations without pixel-level
reconstruction; and a long line of work is pushing towards world models that are good enough to
train real robot policies inside. If Dreamer felt like magic to you, this is the direction in
which that magic scales.

### Robot foundation models beyond π0

In `13_DTVLA.ipynb` you met Decision Transformers and Vision-Language-Action models like π0.
The frontier here moves fast: **π0.5**, **Gemini Robotics**, **Octo**, **OpenVLA**, and the
growing ecosystem of open robot datasets (Open X-Embodiment, DROID) are turning "one policy per
task, per robot" into "one policy for many tasks on many robots". Key open questions: how much
does simulation transfer, how do you combine imitation with online RL fine-tuning without
catastrophic forgetting, and what's the right action representation when the same model drives a
gripper, a humanoid, and a drone?

### Open-endedness and self-play

Some of the most interesting recent work asks what happens when there is *no fixed task* — the
environment itself keeps generating new challenges, and agents improve by training against
ever-better versions of themselves or each other. This is the lineage of **AlphaZero**,
**AlphaStar**, **POET**, and more recently **OMNI-EPIC** and **ADA**. The dream is an agent whose
training curriculum is unbounded: it never runs out of interesting things to learn. If you loved
`11_MCTS.ipynb`'s self-play, this is where that idea leads.

### Continual, safe, and real-world RL

Most algorithms in this course assume a fixed task, a benign environment, and free resets. Real
deployments break all three. **Continual / lifelong RL** asks how an agent can keep learning new
tasks without forgetting old ones. **Safe RL** asks how to enforce hard constraints — the agent
must *never* do X, even during exploration. And the whole subfield of **sim-to-real** is about
closing the gap between the cheap simulator you trained in and the messy physical world you
actually deploy to. These are less glamorous than reasoning LLMs or humanoid robots, but they are
exactly the problems you hit the moment you try to ship an RL system.

## Where to go from here

Reading this list is the easy part. Here's what I'd actually do if I were you, in rough order:

1. **Pick one topic from above and go deep.** Not all of them — one. Read the two or three most
   cited papers, then find an open-source implementation and get it running on your machine.
   Depth beats breadth at this stage of your learning.
2. **Reproduce a result.** Even a small one. Nothing exposes the gap between "I understood the
   paper" and "I understand the paper" like watching your own training curves for the first time.
3. **Break something on purpose.** Take one of the notebooks in this course, change a core
   design choice (remove the target network, swap GAE for plain advantages, turn off the clip in
   PPO), and see what happens. Intuition lives in the ablations.
4. **Build something small and useless.** An RL agent for a silly game, a GRPO fine-tune of a
   tiny LLM on a toy reasoning task, a Dreamer agent in a custom gym env. The goal isn't the
   artefact — it's forcing yourself through the full loop of design, debug, train, evaluate.
5. **Follow the field.** A handful of good sources goes a long way: the
   [r/reinforcementlearning](https://www.reddit.com/r/reinforcementlearning/) subreddit,
   [Lil'Log](https://lilianweng.github.io/), the
   [RL-themed arXiv listings](https://arxiv.org/list/cs.LG/recent), and conference proceedings
   from NeurIPS, ICML, ICLR and CoRL. Pick two, not ten.

## Final words

The journey from "what even is a Markov Decision Process" to "I just trained a world model and
fine-tuned an LLM with GRPO" is not a small one, and if you've done the notebooks honestly — with
the TODOs blank and your own hands on the keys — you should be proud. You've earned the right to
call yourself an RL practitioner.

Keep one thing in mind as you go: the algorithms change, but the fundamentals you've built up in
this course don't. Every new method that shows up on arXiv next year is still going to be about
some combination of *how we estimate value*, *how we update a policy*, *how we explore*, and
*what model of the world we assume*. You now have the vocabulary to read any of it.

Thank you for taking this course. The field is wide open, the tools are in your hands, and the
next breakthrough is as likely to come from a curious learner as from a famous lab — so go chase
the idea that won't leave you alone. 🤖

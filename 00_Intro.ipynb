# Welcome!

Welcome to **Deep Reinforcement Learning: Zero to Hero!** 🚀 This is a hands-on course designed to
take you through the fascinating world of Deep Reinforcement Learning (DRL), from the foundations
all the way to the algorithms powering today's frontier AI systems.

In this series you will write modern RL algorithms from scratch. We'll start with the
[2013 breakthrough](https://arxiv.org/abs/1312.5602) that taught an agent to play Atari games from
pixels, build up through policy gradients to **PPO**, and then branch into the advanced topics
that define the field in 2026: model-based agents that dream inside their own learned world models,
agents that learn from logged data with no environment at all, agents that plan with Monte Carlo
Tree Search, multi-agent systems, and the RL-for-LLMs techniques — **RLHF**, **DPO**, and
**GRPO** — used to train modern reasoning models like OpenAI's o1/o3 and DeepSeek's R1. You'll
even touch Vision-Language-Action models, the class of models driving recent robot foundation
models like π0.

## Why is Reinforcement Learning So Exciting?

Deep Reinforcement Learning can feel less accessible than other areas of AI, but it's also one of
the most intriguing. This is because it tackles a fundamentally different and more autonomous
learning problem: **how can an agent learn to make optimal decisions by interacting with an
environment?**

Unlike other machine learning paradigms, the agent isn't given a dataset of "correct" answers.
Instead, it learns through trial and error, guided only by a sparse **reward signal**. This is
much closer to how humans and animals learn. That unique, interactive learning process is what
lets RL agents reach superhuman performance in games like Go, solve dexterous robotic
manipulation, and — most recently — teach large language models to *reason*, producing the long
chains of thought that define today's frontier models.

### Where Does RL Fit in AI?

AI visionary and Turing Award winner Yann LeCun proposed a powerful analogy to contextualize
different types of learning in AI, which has come to be known as "LeCun's Cake" or "LeCake."

He described the landscape of machine learning as a cake:

- **The Cake**: **Unsupervised (or Self-Supervised) Learning** is the bulk of the cake. It
  represents the vast majority of learning, where models learn the underlying structure of data
  without explicit labels.
- **The Icing**: **Supervised Learning (and Fine-Tuning)** is the icing on the cake. It's a thinner
  layer because it relies on labeled data, which is far less abundant than unlabeled data.
- **The Cherry**: **Reinforcement Learning** is the cherry on top. The reward signal in RL provides
  very little information compared to the rich, high-bandwidth data used in supervised and
  unsupervised learning. As LeCun puts it, it's like "a bit of information per action."

This analogy highlights both the power and the challenge of RL. (A fun footnote: since LeCun
coined this metaphor, the "cherry" has turned out to be more load-bearing than anyone expected —
RL is now one of the core training recipes for frontier LLMs.) While the learning signal is
sparse, the ability to learn from it is the key to creating truly intelligent, autonomous agents
that can navigate the complexities of the real world. This course aims to give you the tools to
build that "cherry."

## What you'll build

The course is split into two halves. The **foundations** take you from first principles up to PPO:

- **MDPs and Tabular RL** — the vocabulary of the whole field
- **DQN** — deep value-based RL, replicating the Atari breakthrough
- **REINFORCE** — the policy-gradient theorem in its purest form
- **Actor-Critic (A2C)** — combining value and policy learning
- **PPO** — the clipped objective that became the modern workhorse

The **advanced** half is a set of self-contained deep dives into the branches of modern RL:

- **Exploration with RND** — solving sparse-reward tasks like `MiniGrid-DoorKey-16x16`
- **Multi-Agent RL** — cooperation, competition, and non-stationarity
- **Offline RL** — learning from logged data with no environment access
- **MCTS & AlphaZero** — planning, search, and self-play
- **RLHF, DPO & GRPO** — the algorithms aligning and teaching reasoning to modern LLMs
- **Decision Transformers & VLAs** — RL as sequence modelling, scaled to robot foundation models
- **Productionising RL** — TensorBoard, Ray, Optuna
- **Model-Based RL & Dreamer** — learning and dreaming inside a world model
- **Meta-RL** — learning to learn, adapting to new tasks in a handful of episodes

The foundations are meant to be done in order; the advanced notebooks are not. Pick whichever
excites you most once you've earned your PPO wings.

## A note on AI-assisted development

The vast majority of this course was written by hand, the old-fashioned way — one notebook at a
time, iterating on explanations, code, and exercises until each section felt right. That's still
the heart of the project, and it's where the pedagogical choices, the structure, and the taste
come from.

In the most recent release, however, I've leaned more heavily on AI assistance. Large language
models were used to **review and refine existing notebooks** — catching subtle bugs, tightening
explanations, suggesting clearer framings — and a **couple of the advanced topics were drafted
with substantial AI help** before being revised, corrected, and stress-tested by hand. Every
notebook still gets run end-to-end, every algorithm still has to actually train, and every
result you see was verified — but it felt only fair to be upfront that some of the prose and
scaffolding in the newer chapters came from a collaboration with the same kind of model you'll
be fine-tuning in `12_RLHF.ipynb`.

If anything, I think this is an honest snapshot of how software and technical content are being
written in 2026: humans still in charge of the vision, the correctness, and the taste, but with
a very capable new collaborator in the loop.

## Prerequisites

To make the most out of this course, you should have the following:

- **Proficiency in Python**: You should be comfortable with Python programming, including concepts
  like classes, data structures (lists, dictionaries), and general control flow.
- **Familiarity with Deep Learning & PyTorch**: You should understand the fundamentals of deep
  learning, such as neural networks, activation functions, and gradient descent. This course uses
  **PyTorch**, so prior experience with it is highly beneficial.
- **Mathematical Foundations (for theory)**: To fully grasp the theory behind the algorithms (which
  is optional but recommended), a solid understanding of basic **calculus** (derivatives), **linear
  algebra** (vectors, matrices), **statistics**, and **probability theory** is required.
